load data

```python
import pandas as pd
df = pd.read_csv("../data/mobile.csv")
df = df.drop("index",axis=1)
df.head(10)
```

```python
df.columns
```

```python
df["price_range"].value_counts()
```

preprocessing (label encoding)

```python
df = df.replace({'yes':1,'no':0,'one': 1, 'dual': 2,'button':0,'touch screen':1,'no wifi':0,'has wifi':1,'medium low':0,'medium high':1,'high':2,'low price':3})

df
```

preprocessing (drop null row and duplicates data)

```python
df = df.dropna()
df
```

```python
df = df.drop_duplicates()
df
```

split feature matrix

```python
X = df[['battery_power', 'blue', 'clock_speed', 'sim type', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g', 'type',
       'wifi']]
X
```

split classification matrix

```python
Y = df["price_range"].values
Y
```

normaliza data

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test = model_selection.train_test_split(X,Y,test_size=0.2,random_state=True)
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
x_test = scaler.fit_transform(x_test)
x_train= scaler.transform(x_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=7).fit(x_train,y_train)
```

```python
yhat=neigh.predict(x_test)
```

evaluation

precision = TP/ (TP + FP)

recall = TP / (TP + FN)

f1_score = 2 * (precision * recall)/(precision + recall)

confusion_matrix:

FP|TN

   0     1     2      3  (predict label)

```python
from sklearn.metrics import accuracy_score , f1_score,confusion_matrix
print(accuracy_score(y_train,neigh.predict(x_train)))
print(accuracy_score(y_test,yhat))
print(f1_score(y_test,yhat, labels=None, pos_label=1, average="micro", sample_weight=None, zero_division='warn'))

from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,yhat)
```
