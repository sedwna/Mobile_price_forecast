```python
import pandas as pd
df = pd.read_csv("../data/mobile.csv")
df = df.drop("index",axis=1)
df.head(10)
```

```python
df.columns
```

```python
df["price_range"].value_counts()
```

```python
df = df.replace({'yes':1,'no':0,'one': 1, 'dual': 2,'button':0,'touch screen':1,'no wifi':0,'has wifi':1,'medium low':0,'medium high':1,'high':2,'low price':3})

df
```

```python
df = df.dropna()
df
```

```python
df = df.drop_duplicates()
df
```

```python
X = df[['battery_power', 'blue', 'clock_speed', 'sim type', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g', 'type',
       'wifi']]
X
X = df.drop(columns='price_range')
```

```python
Y = df["price_range"].values
Y
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test = model_selection.train_test_split(X,Y,test_size=0.2,random_state=True)
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(C=0.01,solver="liblinear").fit(x_train,y_train)
yhat=LR.predict(x_test)
yhat
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_train,LR.predict(x_train)))
print(accuracy_score(y_test,yhat))
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,yhat)
```
