```python
import pandas as pd
df = pd.read_csv("../data/mobile.csv")
df = df.drop("index",axis=1)
df.head(10)
```

```python
df.columns
```

```python
df["price_range"].value_counts()
```

```python
df = df.replace({'yes':1,'no':0,'one': 1, 'dual': 2,'button':0,'touch screen':1,'no wifi':0,'has wifi':1,'medium low':0,'medium high':1,'high':2,'low price':3})

df
```

```python
df = df.dropna()
df
```

```python
df = df.drop_duplicates()
df
```

```python
X = df[['battery_power', 'blue', 'clock_speed', 'sim type', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g', 'type',
       'wifi']]
X
```

```python
Y = df["price_range"].values
Y
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X)
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test = model_selection.train_test_split(X,Y,test_size=0.2,random_state=True)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion="entropy",max_depth=6).fit(x_train,y_train)
predTree=tree.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , f1_score ,confusion_matrix
print(accuracy_score(y_train,tree.predict(x_train)))
print(accuracy_score(y_test,predTree))
print(f1_score(y_test,predTree, labels=None, pos_label=1, average="micro", sample_weight=None, zero_division='warn'))
print(confusion_matrix(y_test,predTree, labels=None, sample_weight=None, normalize=None))
```
